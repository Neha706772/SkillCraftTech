```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url)
```

```python
print(titanic.head())
```

```python
print(titanic.info())
```

```python
print(titanic.describe())
```

```python
print(titanic.isnull().sum())
```

```python
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
```

```python
titanic['Embarked'].fillna(titanic['Embarked'].mode()[0], inplace=True)
```

```python
titanic.drop(columns=['Cabin'], inplace=True)
```

```python
titanic.dropna(subset=['Fare'], inplace=True)
```

```python
titanic['Sex'] = titanic['Sex'].map({'male': 0, 'female': 1})
```

```python
titanic['Embarked'] = titanic['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
```

```python
print(titanic['Survived'].value_counts())
```

```python
plt.figure(figsize=(10, 6))
numeric_titanic = titanic.select_dtypes(include=np.number)
sns.heatmap(numeric_titanic.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic, x='Survived', hue='Sex', palette='viridis')
plt.title("Survival Count by Gender")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(titanic['Age'], kde=True, bins=30, color='blue')
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=titanic, x='Pclass', y='Survived', palette='viridis')
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=titanic, x='Survived', y='Age', palette='viridis')
plt.title("Age vs Survival")
plt.show()
```

```python
sns.pairplot(titanic, hue='Survived', vars=['Age', 'Fare', 'Pclass'], palette='viridis')
plt.show()
```

```python
print("\n--- Insights ---")
print("1. Women had a significantly higher survival rate compared to men.")
print("2. Passengers in higher classes (Pclass 1) had better survival chances.")
print("3. Younger passengers had a higher survival rate compared to older passengers.")
print("4. There is a positive correlation between Fare and Survival, suggesting wealthier passengers were more likely to survive.")
```
